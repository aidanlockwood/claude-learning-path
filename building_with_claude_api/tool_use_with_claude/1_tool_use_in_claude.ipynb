# Tool Use in Claude

By default, Claude can only respond with information it was trained on when it was developed. It cannot return a response on current data and needs to request for information from a server to respond appropriately to the request. 

**Tools** allow Claude to retrieve data to return a response back to the server  

## The Data Flow
1. Initial Request
- User sends a prompt to Claude along with instructions to access relevant data from external sources
2. Tool Request
- Claude analyses the request to determine the tools it needs to use - if it determines it needs additional information
3. Data Retrieval 
- The server executes code to fetch the required information from external APIs/databases
4. Final Response
- Retrieved data returns to Claude, for it to generate a response using original question and newly retrieved data
