# Sending Tool Results

To access the `ToolUseBlock` input argument, locate the object in the response and append `.input`

### Running the ToolResultBlock
- Placed in the content list of the user message
- Communicates the results of running a tool function back to Claude 
- `tool_use_id` - must match the id of the `ToolUse` block that this result corresponds to 
- `content` - Output from running the tool serialised as a string 
- `is_error` - True if an error occurred
