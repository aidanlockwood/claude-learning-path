# Multi Turn Conversations with Tools

```python
from utils.util_funcs import *
from anthropic.types import Message

client, model = api_client_setup()
```

If Claude doesn't request for `ToolUseBlock`, then it is implied that we have reached a final answer to send back to the user

```python
def add_user_message(messages, message): 
    user_message = { 
        "role": "user", 
        "content": message if isinstance(message, Message) else message
    }

    messages.append(user_message)

def add_assistant_message(messages, message): 
    assistant_message = { 
        "role": "assistant", 
        "content": message.content if isinstance(message, Message) else message
    }

    messages.append(assistant_message)
```

```python
messages = []

add_user_message(messages, "What is the current time in HH:MM:SS format?")

response = client.messages.create(
    model = model,
    max_tokens = 1000,
    messages = messages, 
    tools = [get_current_datetime_schema]
)

add_assistant_message(messages, response)

response
```
