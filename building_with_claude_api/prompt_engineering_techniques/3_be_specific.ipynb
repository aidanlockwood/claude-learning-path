# Be Specific

Set specific guidelines to guide the model to be more predictable with its response

***Not Good**<br>
"Write a short story about a character who discovers a hidden
talent"

**Better**<br>
"Write a short story about a character who discovers a hidden
talent
Guidelines:
1. Keep the story under 1,000 words
2. Include a clear action that reveals the character's talent
3. Include at least one supporting character"

This can be achieved through either 
1. List qualities/attributes/structures about what the output should have
2. Provide steps that the model should follow

These may also be combined in production-grade prompts
