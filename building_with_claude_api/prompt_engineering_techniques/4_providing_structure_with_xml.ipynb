# Providing Structure with XML Tags

Handy when including a lot of context
- Considered as delimiters for Claude to understand text structure/groupings

An example 
Here are the last 20 pages of our sales records:<br>
<sales_records> <br>
{sales_records)<br>
</sales_records><br>

This "sales_records" tag is made up, but will help Claude to understand what text corresponds to what
