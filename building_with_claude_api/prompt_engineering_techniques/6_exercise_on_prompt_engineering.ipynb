```python
evaluator = PromptEvaluator(max_concurrent_tasks = 2)

dataset = evaluator.generate_dataset(
    task_description = """
    Extract topics out of a passage of text from a scholarly article into a JSON Array of strings
    """,

    prompt_inputs_spec = { 
        "content": "One paragraph of text from a scholarly journal written in English"
    },
    output_file = "prompt_eng_exercise_data.json",
    num_cases = 4
)
```

```python
def run_prompt(prompt_inputs): 
    prompt = f"""
    At the bottom of this prompt is a number of scholarly articles. Analyse the "content" inside the "prompt_inputs" and determine the topic of which the article belongs to.
    Once you have extracted the input, display it as a JSON array, with the following output. 
    
    <solution_guidelines>
    Do NOT provide any extra commentary about the article.
    Provide a response ONLY with a JSON array of strings, which contain each topic mentioned in the article
    </solution_guidelines>

    Here is an example of a sample input you will receive:
    <text>
    {prompt_inputs['content']}
    </text>

    Follow these steps for your analysis of each article:
    1. Closely Examine the provided text
    2. Identify each topic mentioned
    3. Add each topic to a JSON array
    4. Respond with a JSON array. Do not provide any other text or commentary

    """

    messages = []
    add_user_message(messages, prompt)
    return chat(messages)
```

```python
results = evaluator.run_evaluation(
    run_prompt_function = run_prompt,
    dataset_file = "prompt_eng_exercise_data.json",
    extra_criteria = """
    - Contains a JSON array of strings, containing each topic mentioned in the article.
    - The strings should contain only a topic without any extra commentary
    - Response should contain the JSON array and nothing else
    """
)
```
