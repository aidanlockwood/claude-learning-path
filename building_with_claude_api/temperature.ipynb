# Temperature

Decimal value to dictate how deterministically the output of Claude should be. 

The probability value for this argument is between 0 and 1. 

This value is important for the type of task that is being completed by Claude

- Lower values: 0 -> 0.3
    - Factual Responses
    - Coding assistance
    - Data Extraction

- Medium values: 0.4 -> 0.7 
    - Summarisation
    - Education Content
    - Problem Solving 

- High values: 0.8 -> 1
    - Brainstorming 
    - Creative writing 
    - Joke generation
