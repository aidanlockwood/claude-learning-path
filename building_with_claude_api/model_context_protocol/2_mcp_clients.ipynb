# MCP Clients

Provides a means of communication between the **Client** and **Server**

## MCP Communication

The MCP Specification defines different types of messages that can be exchanged
- `CalToolRequest` 
    - to make a request to an external server
- `CallToolResult` 
    - to get the response from the external server
