# Introduction to MCP
An open standard that allows developers to cut down on the amount of code (by writing tools) they write to implement functionality into an application

An interface to an external service

**Who authors MCP servers?**
Anyone, but typically the service provider offers their own implementation

**How is an MCP Server different from calling the server's API directly?**
MCP Servers provide tools, schemas and implementations. APIs would need the developers to provide these themselves

**Are MCP Servers different from tool use?**
MCP Servers provide tool schemas and functions to be used out of the box
