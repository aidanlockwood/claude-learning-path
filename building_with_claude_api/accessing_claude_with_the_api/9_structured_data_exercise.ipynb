## Creating the Response

Firstly, defining the output configuration for the response

```python
output_config = {
    "format": { 
        "type": "json_schema",
        "schema": {
            "type": "object",
            "properties": {
                "message": {
                    "type": "string"
                },
                "commands": {
                    "type": "array",
                    "items": {
                        "type": "object",
                        "properties": {
                            "name": { 
                                "type": "string"
                            },
                            "command": { 
                                "type": "string"
                            }
                        },
                        "required": ["name", "command"],
                        "additionalProperties": False
                    }
                }
            },
            "required": ["message", "commands"],
            "additionalProperties": False
        }
    }
}
```

Rerunning the prompt with the new `output_config`

```python
messages = []

prompt = """
Generate three different sample AWS CLI commands. Return a JSON object with a short message and three command entries. Each command should be very short.
"""

add_user_message(messages, prompt)
answer = chat(messages, output_config = output_config)

structured = json.loads(answer)
print(structured["message"])
structured["commands"]
```
