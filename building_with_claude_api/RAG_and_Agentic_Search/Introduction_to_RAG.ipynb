# Retrieval Augmented Generation

There are limitations to the amount of text you can include in a prompt. 

If users want to query large documents, it will cost more and take longer to process.

Claude also becomes slightly less effective when dealing with long prompts

### Option 1
Include everything into the prompt

**not the best solution**

### Option 2

Breaking the document into chunks, placing the relevant ones based on the user's question into the prompt


**Upsides**
- Claude only focuses on relevant portions 
- Scales with very large documents
- Technique works with multiple documents
- Smaller prompts: costs less and runs faster

**Downsides**
- Requires a preprocessing step
- Needs a searching mechanism for finding **relevant** chunks 
- Included chunks may not contain all context for Claude to answer the question effectively 
- Many methods to evaluate and chunk the text

**Option 2 is a RAG (Retrieval Augmented Generation) approach**
