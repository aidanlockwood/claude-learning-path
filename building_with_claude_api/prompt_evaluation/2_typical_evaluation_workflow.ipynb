# Typical Evaluation Workflow

## 1. Drafting a Prompt

The goal:
- Writing a prompt to assist users in writing Python code, JSON Config or Regex focused on AWS-specific use cases

- Input
    - User requests code for task
- Output
    - Python, JSON, Regex, with no explanations

## 2. Creating an Evaluation Dataset
Contains typical inputs the system will receive in production. Dataset should include questions that will be interpolated in the prompt template

## 3. Feed through Claude
Take each question/input from dataset, merge with the test prompt and feed through Claude. Retrieve the output responses

## 4. Feed the Claude Response through a Grader

The grader evaluates the quality of the response through examining the original question and Claude's answer. It provides objective scoring, typically from 1 - 10. Higher means better quality

## 5. Modify Prompt and Repeat

Modify the prompt accordingly to ensure it receives the right guidance to help the end user
