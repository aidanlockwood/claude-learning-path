# Prompt Evaluation 
Automated Testing to measure how the prompt works 
- Test against expected answers 
- Compare different versions of the same prompt
- Review outputs for errors

Three different methods exist to approach prompt evaluation

**Option #1** 
Test the prompt once and decide whether it is good enough ❌

**Option #2**
Test the prompt a few times and tweak it to handle edge cases

**Option #3**
Run the prompt through an evaluation pipeline and score it, then iterate the prompt ✅
