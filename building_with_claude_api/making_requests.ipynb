## Creating the API Client

```python
from utils.util_funcs import api_client_setup

client, model = api_client_setup()
```

To make requests to Claude, the `create()` function is used. Taking in the parameters:
- `model`
    - the name of the Claude model being used
- `max_tokens`
    - The maximum number of tokens for the response to output 
    - If Claude responds with more than this argument. The return will only include the length equal to the `max_tokens` argument
- `messages`
    - the input being provided to the model

**Messages Argument**: 
Two types of messages exist for input into Claude
- User Message:
    - Text written by some person (developer/user)
- Assistant Message: 
    - Model-Generated text

```python
# Making a request to the Claude API
message = client.messages.create(
    model = model,
    max_tokens = 1000,
    messages = [{ 
        'role': 'user',
        'content': 'What is quantum computing? Answer in one sentence'
    }]
)
```

Displaying the full text body of the request

```python
message.content
```

Accessing the text component of the response

```python
message.content[0].text
```
