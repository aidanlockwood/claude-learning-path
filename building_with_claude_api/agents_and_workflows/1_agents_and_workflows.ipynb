# Agents and Workflows

## Workflows 
A series of calls to Claude aimed to solve a particular problem through a predetermined series of steps.

**Used when:**
- You can picture the exact flow or steps that Claude should go through to solve a problem
- The UX in your app constrains the user to a set of tasks

## Agents 

Claude is given a goal and a set of tools. Claude is expected to figure out how to complete the goal through the provided tools

**Used when:**
- You're not sure exactly what task or task parameters to give to Claude
