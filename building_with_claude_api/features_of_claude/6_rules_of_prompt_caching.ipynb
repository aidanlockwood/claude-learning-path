# Rules of Prompt Caching

- Initial Request will write the cache
    - Makes the requests cheaper and faster to execute
- Follow up requests can be read from the cache
- Cache lives for up to one hour
- **only useful when repeatedly sending the same content**

Need to manually add a Cache Breakpoint
- anything done before the breakpoint will be cached

Longhand form to cache a request

```
user_message = { 
    "role": "user",
    "content": [
        {
            "type": "text",
            "text": "<long prompt>",
            "cache_control": { 
                "type": "ephemeral"
            }
        }
    ]
}
```

Content after the breakpoint **must** be identical for it to call on the cache.

Breakpoints can be added to tools, schemas and system prompts. This is a useful spot to add breakpoints, as they are usually consistent and don't change significantly

Minimum content length for caching is 1024 tokens. Anything less will not be stored
